```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image1 = cv2.imread("C:/Users/FreeComp/Desktop/open/CV.jpg")
image = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

GaussianBlur = cv2.GaussianBlur(image, (13, 13),0)
bilateral=cv2.bilateralFilter(GaussianBlur,12,51,51)


addWeighted=cv2.addWeighted(bilateral,1.5,np.zeros(bilateral.shape,bilateral.dtype),0,1)

kernal=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharp_image=cv2.filter2D(addWeighted,-1,kernal)

f , ( im1 , im2) = plt.subplots(1 ,2 ,sharey=True ,figsize = (8,5) )
f.subplots_adjust(wspace=.5)
im1.imshow(image ,cmap = "gray" )
im1.set_title("Original_image")
im2.imshow(sharp_image ,cmap = "gray" )
im2.set_title("enhancment_image")
```
